## Script utilizado para balancear o dataset

```python
# Imports
import pandas as pd
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv("data/train.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
# Remove ID
df.drop("ID", axis = 1, inplace = True)
```

```python
df.head()
```

```python
Y = df["TARGET"]
```

```python
X = df.drop(["TARGET"], axis = 1)
```

```python
X.shape, Y.shape
```

```python
df.groupby(["TARGET"]).size()
```

# SMOTE

```python
smote = SMOTE(random_state=42)
```

```python
X_smote, Y_smote = smote.fit_sample(X, Y)
```

```python
X_smote.shape, Y_smote.shape 
```

```python
df_smote = pd.concat([pd.DataFrame(X_smote), pd.DataFrame(Y_smote)], axis = 1)
```

```python
df_smote.columns = df.columns
```

```python
df_smote.head()
```

```python
df_smote.shape
```

```python
df_smote.groupby("TARGET").size()
```

```python
df_smote.to_csv("data/clientes_smote.csv", index=False)
```
